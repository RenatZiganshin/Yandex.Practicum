# Исследование объявлений о продаже квартир

Нужно научиться определять рыночную стоимость объектов недвижимости. Задача — установить параметры. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма.

## Предобработка данных

### Удаление пропусков

Пропуски есть во многих столбцах. Рассмотрим их по порядку.

#### Высота потолков

В столбце 'ceiling_height' (высота потолков) нет чуть менее половины всех значений. Возможно, авторы объявлений не знали точной высоты потолков или просто пропустили этот пункт при размещении объявлений. Заполнение пропусков медианным или средним арифметическим значением сильно исказит результаты дальнейшего анализа, так как это существенная часть всех значений, поэтому оставим пропуски.



#### Количество этажей

В столбце 'floors_total' (всего этажей в доме) заполнено 23613 из 23699 значений. Оценим диапазон и распределение значений, чтобы выбрать значение для заполнения пропусков.

```python
print('Минимальное количество этажей:', data['floors_total'].min())
print('Максимальное количество этажей:', data['floors_total'].max())
print('Среднее арифметическое количество этажей:', data['floors_total'].mean())
print('Медианное количество этажей:', data['floors_total'].median())
data.hist(column='floors_total')
```

Наибольшее количество значений находятся в диапазоне 1-10, медианное значение 9,0, среднее арифметическое 10,67. Заполним пропуски медианой.

```python
data['floors_total'] = data['floors_total'].fillna(9)
```

#### Жилая площадь

Аналогично предыдущему столбцу оценим значения в столбце 'living_area' (жилая площадь в квадратных метрах), в котором пропущенные значения тоже составляют малую часть.

```python
print('Минимальная площадь:', data['living_area'].min())
print('Максимальная площадь:', data['living_area'].max())
print('Среднее арифметическое площади:', data['living_area'].mean())
print('Медианная площадь', data['living_area'].median())
data.hist(column='living_area')
```

Подавляющее большинство квартир имеют площадь менее 50 м2. Заполним пропуски медианным значением.

```python
data['living_area'] = data['living_area'].fillna(30)
```

#### Аппартаменты

Следующий столбец с пропусками - 'is_apartment' (булев массив, указывающий на то, является ли жилье апартаментами). Посмотрим, какие значения и в каком количестве он содержит.

```python
print(data['is_apartment'].value_counts())
```

Из 23699 объявлений в 2725 указано, что это не апартаменты, а в 50 указано, что апартаменты. Так как это важная характеристика недвижимости, в объявлениях о продаже апартаментов это обязательно бы указали. Поэтому отсутствие значений в столбце 'is_apartment' указывает на то, что продаваемая квартира не является апартаментами. Заполним пропуски соответствующим значением.

```python
data['is_apartment'] = data['is_apartment'].fillna(False)
```

#### Площадь кухни

Перед заполнением пропусков в столбце 'kitchen_area' (площадь кухни в квадратных метрах) рассмотрим столбец 'open_plan' (свободная планировка). Возможно, пропуски в площади кухни связаны с тем, что планировка квартиры еще не определена.

```python
print(data['open_plan'].value_counts())
```

Также причиной отсутствия площади кухни может быть то, что это объявление о продаже студии, в которой кухня объединена с основной комнатой. Посчитаем количество студий.

```python
print(data['studio'].value_counts())
```

Свободная планировка всего лишь у 67 квартир, а 149 квартир являются студиями, в то время как пропуски в площади кухни имеются более чем в 2000 обявляений. Можем заполнить пропуски в квартирах со свободной планировкой и в студиях таким же образом как и остальные пропуски. Рассмотрим столбец 'kitchen_area' (площадь кухни в квадратных метрах)

```python
print('Минимальная площадь кухни:', data['kitchen_area'].min())
print('Максимальная площадь кухни:', data['kitchen_area'].max())
print('Среднее арифметическое площади кухни:', data['kitchen_area'].mean())
print('Медианная площадь кухни', data['kitchen_area'].median())
data.hist(column='kitchen_area')
```

Заполним пропуски медианным значением

```python
data['kitchen_area'] = data['kitchen_area'].fillna(9.1)
```

#### Число балконов

Cледующий столбец с пропусками - 'balcony' (число балконов). Предположим, что если в объявлении не указано число балконов, то их нет в квартире. Заменим пропуски на 0.

```python
data['balcony'] = data['balcony'].fillna(0)
```

#### Населенный пункт

Следующий столбец с пропусками - 'locality_name' (название населённого пункта). Пропущенных значений в нем всего лишь 49 из 23699. Посмотрим, какие населенные пункты представлены в нашем датафрейме

```python
print(data['locality_name'].value_counts())
```

Отмечается абсолютное преобладание Санкт-Петербурга в объявлениях (15721 объявление), при этом на втором месте по популярности посёлок Мурино с всего лишь 522 объявлениями. Чтобы заполнить небольшое количество пропусков и не потерять адекватность данных, предположим, что это объявления из Санкт-Петербурга.

```python
data['locality_name'] = data['locality_name'].fillna('Санкт-Петербург')
```

### Анализ квартир в центре

#### Площадь, цена, число комнат и высота потолков

Создадим сводную таблицу, в которой сравним среднее и медианное значения площади, цены, числа комнат и высоты потолков в центре и за его пределами. Для этого сперва создадим булев столбец 'centre', где True - это месторонахождение квартиры в радиусе 7 км от центра.

```python
#функция для одной строки
def add_centre(row):
    if row['cityCenters_nearest_km'] <= 7 and row['locality_name'] == 'Санкт-Петербург':
        return True
    else:
        return False
#примеим функцию к датафрейму
data['centre'] = data.apply(add_centre, axis=1)
#построим сводную таблицу
centre_stat = data.pivot_table(columns='centre', values=('total_area', 'last_price', 'rooms', 'ceiling_height'), aggfunc={'mean', 'median'})
centre_stat.columns = ['Не центр', 'Центр']
centre_stat
```

**Из таблицы видно, что при неизменной высоте потолков у квартир в центре больше комнат и, соответственно, больше площадь. Логично, что и стоимость квартир в центре выше.**

#### Факторы, влияющие на стоимость квартир

#####  Количество комнат

Определим вляние количества комнат на стоимость квартиры аналогично тому, как это делали ранее со всеми квартирами

```python
display(data.query('centre == True').pivot_table(index='rooms', values='last_price', aggfunc={'mean', 'median'}))
data.query('centre == True').pivot_table(index='rooms', values='last_price', aggfunc={'mean', 'median'}).plot()
plt.title('Стоимость квартиры от количества комнат')
plt.show()
```

**Как и следовало ожидать, чем больше комнат в квартире, тем она дороже**

##### Этаж

Построим столбчатую диаграмму зависимости стоимости квартиры от того, на каком этаже она находится (первом, последнем или другом)

```python
display(data.query('centre == True').pivot_table(index='floor_cat', values='last_price', aggfunc={'mean', 'median'}))
data.query('centre == True').pivot_table(index='floor_cat', values='last_price', aggfunc={'mean', 'median'}).plot(kind='bar')
plt.title('Стоимость квартиры от этажа')
plt.show()
```

**Квартиры на первом этаже самые дешевые, на последнем подороже, но самые дорогие все-таки на промежуточных этажах.**

##### Дата размещения объявления

Построим столбчатые диаграммы и сводные таблицы с зависимостью стоимости квартиры в центре от дня недели, месяца и года размещения объявления.

```python
print('Месяц')
display(data.query('centre == True').pivot_table(index='month', values='last_price', aggfunc={'mean', 'median'}))
data.query('centre == True').pivot_table(index='month', values='last_price', aggfunc={'mean', 'median'}).plot(kind='bar')
plt.title('Стоимость квартиры от месяца размещения объявления')
plt.show()
```

```python
print('День недели')
display(data.query('centre == True').pivot_table(index='weekday', values='last_price', aggfunc={'mean', 'median'}))
data.query('centre == True').pivot_table(index='weekday', values='last_price', aggfunc={'mean', 'median'}).plot(kind='bar')
plt.title('Стоимость квартиры от дня недели размещения объявления')
plt.show()
```

```python
print('Год')
display(data.query('centre == True').pivot_table(index='year', values='last_price', aggfunc={'mean', 'median'}))
data.query('centre == True').pivot_table(index='year', values='last_price', aggfunc={'mean', 'median'}).plot(kind='bar')
plt.title('Стоимость квартиры от года размещения объявления')
plt.show()
```

**Видно, что стоимость квартиры не зависит от дня недели, месяца и года размещения объявления.**

##### Вывод

**Можно отметить, что стоимость квартир в центра Санкт-Петербурга зависит от тех же факторов, что и все квартиры в округе. За исключением одного - общее падение цен на квартиру, которое было отмечено в 2016-2017 по всему датафрейму, не задело центральные квартиры Санкт-Петербурга.**
